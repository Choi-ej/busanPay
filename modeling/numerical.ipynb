```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("수치형합본.csv",encoding='cp949',low_memory =False)
```

```python
df
```

```python
data = df.iloc[:,1:]
data
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, 
            init='random', 
            n_init=10, 
            max_iter=100,
            tol=1e-04,
            random_state=0)
```

```python
original = km.fit_predict(data)
```

```python
original
```

```python
df['class1'] = original
```

```python
df.loc[:, ['시군구명','class1']]
```

```python
#original data

distortions = [ ]
for i in range(1, 11):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=10, 
                max_iter=300, 
                random_state=0)
    km.fit(data)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
data_s = scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, 
            init='random', 
            n_init=10, 
            max_iter=100,
            tol=1e-04,
            random_state=0)
```

```python
class_std = km.fit_predict(data_s)
class_std
```

```python
df['class_std'] = class_std
```

```python
from sklearn.decomposition import PCA

pca = PCA()
data_pcs = pca.fit_transform(data_s) #정규화된 데이터를 사용
pca.explained_variance_ratio_
```
