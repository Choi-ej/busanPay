```python
store = pd.read_csv("부산광역시_지역화폐(동백전) 가맹점 현황_20221222.csv",encoding = 'cp949', low_memory =False)
store.head()
```

```python
len(store)
```

```python
store = store.drop(['순번'], axis = 1)
```

```python
# 도로명 대신 지번 채워진 결측치 합쳐서 주소 컬럼 생성
store['주소'] = store['도로명']
store['주소'] = store['도로명'].fillna(store['지번'])
```

```python
# 결측치 채워졌나 확인
store['주소'].isnull().sum()
```

```python
# 타지역 사업자등록번호 때문에 부산 주소만
store = store[store['주소'].str.startswith('부산')]
```

```python
len(store) # 타 지역 40개 제거
```

```python
store = store.reset_index(drop=True) #제거된 행 인덱스 리셋
```

```python
# 부산광역시를 부산으로 수정.
store_name = store['주소'].str.replace('부산광역시','부산')
# 부산시를 부산으로 수정
store_name = store['주소'].str.replace('부산시','부산')
```

```python
store_name
```

```python
# 부산광역시 -> 부산 변경한 컬럼 데이터 넣기
store['주소'] = store_name
```

```python
# 인덱스 리셋
store = store.reset_index(drop=True)
store
```

```python
store.loc[12850, :]
```

```python
# 시군구명 추출
for i in range(0,len(store)):
    store.loc[i, '시군구명'] = store.loc[i, '주소'][3:store.loc[i,'주소'].find(' ', 3)]
```

```python
# 빈 값 확인 (결측치는 아니고 빈칸이 많아서 입력 오류된 데이터)
store['시군구명'].sort_values()
```
