```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("수치형합본.csv",encoding='cp949',low_memory =False)
```

```python
df
```

```python
data = df.iloc[:,1:-2]
data
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, 
            init='random', 
            n_init=10, 
            max_iter=100,
            tol=1e-04,
            random_state=0)
```

```python
y_km = km.fit_predict(data)
```

```python
y_km
```

```python
df['class'] = y_km
```

```python
df.iloc[:,[0,-1]]
```

```python
distortions = [ ]
for i in range(1, 11):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=10, 
                max_iter=300, 
                random_state=0)
    km.fit(data)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
data_s = scaler.fit_transform(data)
```

```python
y_s = km.fit_predict(data_s)
y_s
```

```python
df['class_s'] = y_s
df.iloc[:,[0,-2,-1]]
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy', 
                              max_depth=5, 
                              random_state=1)
tree.fit(data, y_km)
```

```python
f_names = list(data.columns)
```

```python
f_names
```

```python
class_names = np.unique(y_km)
class_names
```
